```r
# Adjusts range
# Returns 2 values 
# First value is the range adjusted with decimal values (5 -> 4.83%) 
# Second value is the hands belonging to that range (1 -> 2 (AA, KK); 100 -> 169)

# We use the order of hands from the PokerStove software
# Adjusts to the nearest value in PORC2

rajust <- function(rango)
{
  rango <- rango / 100
  .rajust <- c(0.004524887, 1)
  for (i in 2:169)
  {
    if (rango >= PORC2[i - 1, 1] & rango < PORC2[i, 1])
    {
      if (rango - PORC2[i - 1, 1] <= PORC2[i, 1] - rango)
      {
        .rajust <- c(PORC2[i - 1, 1], i - 1)
      } else
        if (rango - PORC2[i - 1, 1] > PORC2[i, 1] - rango)
        {
          .rajust <- c(PORC2[i, 1], i)
        }
    }
  }
  return(.rajust)
}
```

```r
# Returns the minimum range against which a hand has a specific equity
Eq.Min <- function(mano, equity)
{
  temp <-
    170 - length(subset(manos_rangos[num.mano(mano),], manos_rangos[num.mano(mano),] > equity / 100))
  return(100 * PORC2[temp,])
}
```

```r
# Example
Eq.Min('JTo', 40)
Eq.Min('JTo', 50)
```

```r
# Creates a table with the minimum values from Eq.Min for all the hands 
Eq.Min.Tabla <- function(equity)
{
  a <- matrix(rep(0, 169 * 2), 169, 2)
  
  for (i in 2:170)
  {
    a[i - 1, 1] <- names(PF)[i]
    a[i - 1, 2] <- Eq.Min(names(PF)[i], equity)
  }
  return(a)
}
```

```r
# EQ hand vs range
EQ <-
  function(mano, rango1, rango2 = 0)
    # rango1 es el total and rango2 is the one substracted at the beginning of the range for 3 Betting.
  {
    if (rango1 > 100)
    {
      rango1 <- 100
    }
    if (rango2 <= 0)
    {
      eq <- manos_rangos[num.mano(mano), rajust(rango1)[2]]
    } else
      if (rango2 > 0)
      {
        rango <- c((rajust(rango2)[2] + 1):rajust(rango1)[2])
        # rango <- c(1:rajust(rango1)[2])
        # Vector de manos que entran en el rango al que nos enfrentamos
        
        EQS <- rep(0, rajust(rango1)[2])
        PORCs <- rep(0, rajust(rango1)[2])
        PORCs <- (PORC[1:rajust(rango1)[2], 1])
        #/(PORC[rajust(rango1)[2], 1]) Porcentaje individual de cada una de las manos, sin acumular
        
        # A partir de aquí modifica los porcentajes de cada mano en función de la mano que tenga el otro
        # V5: 1 pareja, 2 off suited, 3 suited
        
        t1 <- subset(pars, pars[, 2] == mano)
        # t1 es la mano en la matriz pars
        
        t2 <- pars[1:rajust(rango1)[2],]
        # t2 es el rango total al que nos enfrentamos en la matriz pars
        
        t3 <-
          subset(t2, t2[, 3] == t1[, 3] |
                   t2[, 3] == t1[, 4] | t2[, 4] == t1[, 3] | t2[, 4] == t1[, 4])
        # t3 es la parte del rango al que nos enfrentamos con alguna carta coincidente con nuestra mano
        
        if (t1[5] == 1)
          # Si nuestra mano es una pareja
        {
          for (i in t3[, 1])
          {
            if (t2[i, 5] == 1)
            {
              PORCs[i] <- PORCs[i] / 6
            }
            if (t2[i, 5] == 2 | t2[i, 5] == 3)
            {
              PORCs[i] <- PORCs[i] / 2
            }
          }
        } else
          if (t1[5] == 2)
            # Si nuestra mano es off suited
          {
            for (i in t3[, 1])
            {
              if (t2[i, 5] == 2 & t2[i, 3] == t1[3]  &  t2[i, 4] == t1[4])
              {
                PORCs[i] <- PORCs[i] * 7 / 12
              }
              if (t2[i, 5] == 2 &
                  (t2[i, 3] != t1[3]  |  t2[i, 4] != t1[4]))
              {
                PORCs[i] <- PORCs[i] * 3 / 4
              }
              if (t2[i, 5] == 1)
              {
                PORCs[i] <- PORCs[i] / 2
              }
              if (t2[i, 5] == 3 &
                  t2[i, 3] == t1[3]  &  t2[i, 4] == t1[4])
              {
                PORCs[i] <- PORCs[i] / 2
              }
              if (t2[i, 5] == 3 &
                  (t2[i, 3] != t1[3]  |  t2[i, 4] != t1[4]))
              {
                PORCs[i] <- PORCs[i] * 3 / 4
              }
            }
          } else
            if (t1[5] == 3)
              # Si la mano es suited
            {
              for (i in t3[, 1])
              {
                if (t2[i, 5] == 2 & t2[i, 3] == t1[3]  &  t2[i, 4] == t1[4])
                {
                  PORCs[i] <- PORCs[i] * 1 / 2
                }
                if (t2[i, 5] == 2 &
                    (t2[i, 3] != t1[3]  |  t2[i, 4] != t1[4]))
                {
                  PORCs[i] <- PORCs[i] * 3 / 4
                }
                if (t2[i, 5] == 1)
                {
                  PORCs[i] <- PORCs[i] / 2
                }
                if (t2[i, 5] == 3 &
                    t2[i, 3] == t1[3]  &  t2[i, 4] == t1[4])
                {
                  PORCs[i] <- PORCs[i] * 3 / 4
                }
                if (t2[i, 5] == 3 &
                    (t2[i, 3] != t1[3]  |  t2[i, 4] != t1[4]))
                {
                  PORCs[i] <- PORCs[i] * 3 / 4
                }
              }
            }
        t4 <- PORCs[1]
        t5 <- sum(PORCs[rango])
        
        for (i in rango)
        {
          EQS[i] <- PF[, mano][i]
          # PF[, mano] devuelve los eqs de nuestra mano contra todas las demás
          
          PORCs[i] <- (PORCs[i]) / (t5)
        }
        eq.temp <- EQS * PORCs
        #eq.temp <- eq.temp[(rajust(rango2)[2] + 1):rajust(rango1)[2]]
        eq.temp <- eq.temp[1:rajust(rango1)[2]]
        
        eq <- sum(eq.temp)
      }
    return(eq)
  }
```

```r
X3B10BBs <- function(mano)
{
  # Eq.Min(hand, equity) is the minimum range against which a hand has a specific equity
  # Assuming there is no fold to 3B.
  # In BB we must pay 9 BB for a pot of 20.5, that makes 9/20.5 = 0.44 EQ needed.
  # In SB the pot has ((EQ(mano, 5) * 22) - 9.5) less chips 5% of the time.
  # EQ(mano, 5) is the overcall after we've called with our hand. Estimted 5% range.
    
  # SB, we pay 9.5 for a pot of 21.
  # BB cost is (EQ(mano, 5) * 22) - 9.5

  # SB
  a <-
    Eq.Min(mano, 100 * (9.5 - ((EQ(mano, 5) * 22) - 9.5) * 0.05)/21)

  # BTN
  b <-
    Eq.Min(mano, 100 * (10 - ((EQ(mano, 5) * 22) - 10) * 0.1)/21)
    
  # CO
  c <-
    Eq.Min(mano, 100 * (10 - ((EQ(mano, 5) * 22) - 10) * 0.15)/21)
    
  # MP
  d <-
    Eq.Min(mano, 100 * (10 - ((EQ(mano, 5) * 22) - 10) * 0.2)/21)
  
  return(c(a, b, c, d))
}
```

```r
###########################
###########################

# 3B no All In

############################
############################

# Fórmula básica
# EV = %Fold * beneficio por Fold + %Call * ((Tamaño del bote * Equity) - Coste de entrar en el bote)

# mano_ <- nuestra mano
# CC <- Rango de cold call de los villanos por detrás
# pos <- nuestra posición según el código:
# pos = 1, BB; pos = 2, SB; pos = 3, D; pos = 4, CO; pos = 5, MP; pos = 6, EP con poster
# stack <- nuestro stack sin contar la ciega
# La primera columna del resultado es el porcentaje de fold de su limp

# Tabla de 3B contra rivales que sólo hacen 4B ó fold
# OR = 3BBs, raiser D o antes, 3B a 7 BBs, cold 4B = 5, foldeamos siempre ante C4B.
# TOR tamaño del OR del villano
# Hay que incluir si hacemos call al open 4B
# Comprobar cual es el CC típico en el HM y rehacer las tablas con eso
# Incluir F3B = 0
# El tamaño del 3B no incluye la ciega puesta.
# Para seleccionar el peor CC ponemos CC = -1

X3B <-
  function(mano_,
           OR,
           pos,
           stack,
           TOR,
           T3B,
           allin,
           CC,
           C3B,
           rake1 = 0.05,
           rake2 = 3)
    #pos = 1, BB; pos = 2, SB; pos = 3, D; pos = 4, CO; pos = 5, MP
  {
    # Establece el rake mediante una función
    
    raked <- function(x)
    {
      temp <- rake1 * x
      if (temp > rake2) {
        temp <- rake2
      }
      return(x - temp)
    }
    
    if (allin)
    {
      # Se crea la estructura de la matriz de resultados
      
      a <- seq(0, 95, 5)
      cc <- matrix(0, length(a), 3)
      colnames(cc) <- c('F3B', 'won 3B/push', 'OR')
      cc[, 1] <- a
      cc[, 3] <- rep(OR, length(a))
      
      # Establece el peor cold 4B (CC) para nuestra mano
      
      if (pos == 1) {
        CC <- 0
      } else
        
        if (CC == -1)
        {
          perd.temp <- rep(0, 100)
          for (i in 1:100)
          {
            perd.temp[i] <-
              (i / 100) * (pos - 1) * (-(stack - EQ(mano_, i) * raked(2 * stack + 0.75 + TOR)))
          }
          # Almacenar el EQ que acabamos de calcular para que vaya más rápido luego
          # Después de unos cuantos que no empeore parar el bucle
          
          ss <- min(perd.temp)
          CC <- which(perd.temp == ss)
        }
      
      # Establecemos perd, que es lo que perdemos / ganamos por cada jugador que queda por hablar cuando nos hace call.
      
      perd <- -(stack - EQ(mano_, CC) * raked(2 * stack + TOR + 1))
      # El dinero muerto cambia en función de si el open 4B viene
      # de las ciegas o no. Calculamos una media de 1 + ó -.
      
      # B2 <- bote que se forma con el push - call.
      
      if (pos > 2)
      {
        B2 <- raked(2 * stack + 1.5)
      } else
        
        if (pos == 2)
        {
          B2 <- raked(2 * stack + 2)
        } else
          
          if (pos == 1)
          {
            B2 <- raked(2 * stack + 2.5)
          }
      
      # Calculamos la matriz
      
      for (i in a)
        # a son los F3B de 5 en 5
      {
        # Proporción de fold al 3B
        fold <- i / 100
        
        # E2 <- equity de nuestra mano contra el rango de call del villano
        E2 <- EQ(mano_, OR * (1 - fold))
        
        # Ganancia por fold equity
        FE <- fold * (TOR + 1.5)
        
        # Ganancia 3B / push
        cc[1 + (i / 5), 2] <-
          (CC / 100) * (pos - 1) * (perd) + (1 - ((CC / 100) * (pos - 1))) * (FE + (1 - fold) * ((B2 * E2) - stack))
      }
      
      # Devuelve la matriz cc
      return(cc)
      
    } else
      
      if (allin == F)
      {
        # Se crea la estructura de la matriz de resultados
        
        a <- seq(0, 95, 5)
        cc <- matrix(0, length(a), 5)
        colnames(cc) <-
          c('F3B', 'C3B', 'won 3B/push', 'won 3B/F', 'OR')
        cc[, 1] <- a
        cc[, 2] <- C3B
        cc[, 5] <- rep(OR, length(a))
        
        # Establece el peor CC para nuestra mano cuando es allin
        # Está pendiente cuando no es allin
        
        if (pos == 1)
        {
          CC <- 0
          ss <- 0
        }
        
        if (CC == -1)
        {
          perd.temp <- rep(0, 100)
          for (i in 1:100)
          {
            perd.temp[i] <-
              (i / 100) * (pos - 1) * (-(stack - EQ(mano_, i) * raked(2 * stack + 2)))
          }
          # Almacenar el EQ que acabamos de calcular para que vaya más rápido luego
          # Después de unos cuantos que no empeore parar el bucle
          
          perd.temp.2 <- c(1:100) * (pos - 1) * (-T3B) / 100
          
          sss <- cbind(perd.temp, perd.temp.2)
          colnames(sss) <- c('3B / push', '3B / fold')
          
          sss2 <- NULL
          
          for (i in 1:100)
          {
            sss2 <- c(sss2, max(sss[i,]))
          }
          
          ss <- min(sss2)
          CC <- which(sss2 == ss)
        }
        
        # Proporción de call al 3B
        
        call_ <- C3B / 100
        
        for (i in a)
        {
          # Proporción de fold al 3B
          fold <- i / 100
          
          # Proporción de 4B
          push <- (100 - i - C3B) / 100
          
          # E1 <- equity de nuestra mano contra el rango de push del villano
          E1 <- EQ(mano_, OR * push)
          
          # E2 <- equity de nuestra mano contra el rango de call del villano
          if (call_ > 0)
          {
            E2 <- EQ(mano_, OR - OR * fold, OR * push)
          } else
            
            if (call_ == 0)
            {
              E2 <- 0
            }
          
          # C2 <- coste del movimiento 3B / fold
          # B2 <- bote que se forma con el 3B / push
          # B1 <- bote que se forma con el 3B - call
          if (pos > 2)
          {
            # Esto si es push
            C2 <- T3B * push + (pos - 1) * (CC / 100) * T3B
            
            B2 <- raked(2 * stack + 1.5)
            
            # Esto si es 3b - call
            
            B1 <- raked(2 * T3B + 1.5)
          } else
            
            if (pos == 2)
            {
              # Esto si es push
              C2 <- T3B * push + (CC / 100) * T3B
              
              B2 <- raked(2 * stack + 2)
              
              # Esto si es 3b - call
              
              B1 <- raked(2 * T3B + 2)
            } else
              
              if (pos == 1)
              {
                # Esto si es push
                C2 <- T3B * push
                
                B2 <- raked(2 * stack + 2.5)
                
                # Esto si es 3b - call
                
                B1 <- raked(2 * T3B + 2.5)
              }
          
          # Ganancia por fold equity
          FE <- fold * (TOR + 1.5)
          
          # Showdown EQ botes call
          SE <- call_ * B1 * E2
          
          # Ganancia 3B / push
          cc[1 + (i / 5), 3] <-
            FE + push * ((B2 * E1) - stack) + ss + SE
          
          # Ganancia 3B / fold
          cc[1 + (i / 5), 4] <- FE - C2 + SE
        }
        return(cc)
      }
  }
```

```r
# Example:
mano_ <- 'X55'
OR <- 40
pos <- 1 # BB
stack <- 20
TOR <- 2 # OR 2 bb
T3B <- 20 # 3B 20 bbs
allin <- T
CC <- -1
C3B <- 0
rake1 <- 0.05
rake2 <- 3

X3B(mano_, OR, pos, stack, TOR, T3B, allin, CC, C3B, rake1 = 0.05, rake2 = 3)

# Opens 40 and optimal F3b For villain is 10% (min won for us)
```

```r
Eq.Min.Rango <- function(rango, equity)
{
  orden_educa()
  equity <- equity / 100
    
  # matrix of zeros
  # It will be a matrix of every hand against that range
  d <- matrix(0, 169, 1)
  
  for (i in 1:169)
  {
    d[i, 1] <- EQ(names(PF)[-1][i], rango)
  }
  
  b <- 0
  
  for (i in 1:nrow(d))
  {
    if (d[i, 1] > equity)
    {
      b <- b + PORC[i, 1]
    }
  }
  return(b)
}
```

```r
# Against a range of 50, almost every hand has 30% eq, so the range is almost 100%

Eq.Min.Rango(50, 30)
EQ('X72o', 50)
```
